# STEP 3- ANALYZIE THE DATA TO SOLVE THE QUESTIONS

# 1.Show the number of Confirmed, Deaths and Recovered cases in each Region


```python
data=df.groupby('Region')['Confirmed','Deaths','Recovered'].sum()
```

```python
print(data)
```

# 2.Remove all the Records where Confirmed Cases are Less than 10

```python
df=data1[~(data1['Confirmed']<10)]
df
```

# 3.In Which Region their maximum number of Confimred cases were recorded

```python
df.groupby('Region').max('Confirmed').sort_values(by='Confirmed',ascending=False)
```

# 4.In which Region maximum number of Deaths cases were recorded


```python
df.groupby('Region').max('Deaths').sort_values(by='Deaths',ascending=False).head(5)
```

# 5.How many Confirmed, Deaths & Recovered cases were reported from India till 29 April 2020 

```python
df[df['Region']=='India']
```

# 6-A ) Sort the entire data wrt No. of Confirmed cases in ascending order.

```python
df.sort_values(by='Confirmed',ascending=True).head(25)
```

# 6-B) Sort the entire data wrt No. of Confirmed cases in decending order.

```python
df.sort_values(by='Confirmed',ascending=False).head(25)
```
